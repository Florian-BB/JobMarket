ADZUNA utilise des catégories, je les récupère dans un premier temps

```python
import requests
import json
import os

# mettre son appID et sa key puis supprimer.
appid = "."
key = "."
url = "http://api.adzuna.com/v1/api/jobs/fr/categories?app_id={0}&app_key={1}&&content-type=application/json".format(appid, key)
req1 = requests.get(url).json()

categories = req1["results"]


tags = [] # j'initie cette liste qui récupèrera les tags qui correspondent aux catégories

for cat in categories:
    tags.append(cat["tag"]) # Pour chaque catégorie je récupère son tag

print(tags)


```

Recherches générales

```python
results, success, failed = [], [], [] # Création de 3 listes, une pour les résultats, une pour la journalisation des succés et des échecs

for page in range(1, 2):
    url = "https://api.adzuna.com/v1/api/jobs/fr/search/{0}?app_id={1}&app_key={2}".format(page, appid, key)
    req = requests.get(url)
    if req.status_code == 200: # Si réponse positive, alors on récupère les données
        try:
            data = req.json()
            results.append(data["results"])
        except:
            failed.append("{} : {}".format(page))
            print(req.status_code, " pour la page ", page)
        else:
            print(req.status_code, " pour la page ", page)
            success.append("{} : {}".format(page))

    else :
        print(req.status_code, " pour la page ", page)
```

Recherche par catégories

```python
results, success, failed = [], [], [] # Création de 3 listes, une pour les résultats, une pour la journalisation des succés et des échecs
for tag in tags:
    for page in range(1, 2):
        url = "https://api.adzuna.com/v1/api/jobs/fr/search/{0}?app_id={1}&app_key={2}&category={3}".format(page, appid, key, tag)
        req = requests.get(url)
        if req.status_code == 200: # Si réponse positive, alors on récupère les données
            try:
                data = req.json()
                results.append(data["results"])
            except:
                failed.append("{} : {}".format(tag, page))
                print(req.status_code, " pour la page ", page, "de la catégorie ", tag)
            else:
                print(req.status_code, " pour la page ", page, "de la catégorie ", tag)
                success.append("{} : {}".format(tag, page))

        else :
            print(req.status_code, " pour la page ", page, "de la catégorie ", tag)
```

Recherche par pays

```python
countries = ["gb", "fr", "us","at","au","be","br","ca","ch"]
results, success, failed = [], [], [] # Création de 3 listes, une pour les résultats, une pour la journalisation des succés et des échecs
for country in countries:
    for page in range(1, 11):
        url = "https://api.adzuna.com/v1/api/jobs/{3}/search/{0}?app_id={1}&app_key={2}".format(page, appid, key, country)
        req = requests.get(url)
        if req.status_code == 200: # Si réponse positive, alors on récupère les données
            try:
                data = req.json()
                results.append(data["results"])
            except:
                failed.append("{} : {}".format(country, page))
                print(req.status_code, " pour la page ", page, "du pays ", country)
            else:
                print(req.status_code, " pour la page ", page, "du pays ", country)
                success.append("{} : {}".format(country, page))

        else :
            print(req.status_code, " pour la page ", page, "du pays ", country)


```

Recherche des top companies

```python
countries = ["gb", "fr", "us","at","au","be","br","ca","ch"]
top_companies = [] # Création de 3 listes, une pour les résultats, une pour la journalisation des succés et des échecs
for country in countries:
    url = "https://api.adzuna.com/v1/api/jobs/{2}/top_companies?app_id={0}&app_key={1}".format(appid, key, country)
    req = requests.get(url)
    if req.status_code == 200: # Si réponse positive, alors on récupère les données
        try:
            data = req.json()
            top_companies.append(data)
        except:
            print(req.status_code, " pour le pays ", country)
        else:
            print(req.status_code, " pour le pays ", country)

    else :
        print(req.status_code, " pour le pays ", country)


```

Affichage des données

```python

print(len(results))
print(len(top_companies))

for page in results:
    for r in page:
        print("\nTitre du poste : ", r['title'])
        print("Nom de la société : ", r['company']['display_name'])
        print("Pays : ", r['location']['area'][0])
    


```

Enregistrement des données dans un fichier

```python
with open("adzuna_results.json", "w") as outfile:
    json.dump(results, outfile)

with open("adzuna_top_companies.json", "w") as outfile:
    json.dump(top_companies, outfile)

with open("adzuna_categories.json", "w") as outfile:
    json.dump(categories, outfile)
```
